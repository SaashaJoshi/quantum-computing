```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
plt.figure()
axis = plt.gca()    # Plots axes
axis.quiver([2], [3], angles = 'xy', scale_units = 'xy', scale = 1)     # Plot 2D arrows
axis.set_xlim([0, 5])
axis.set_ylim([0, 5])
plt.draw()
plt.show()
```

```python
plot_bloch_vector([1, 0, 0])    # Superposition between  |0  and  |1
#  Each of the points on the surface of the sphere represents a different quantum state.
```
