```python
import numpy as np
from qiskit import *
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
# Creating a Bell Pair
def bell_pair(circuit, qr1, qr2):
    circuit.h(qr1)
    circuit.cx(qr1, qr2)
```

```python
qr = QuantumRegister(3)
crz = ClassicalRegister(1)
crx = ClassicalRegister(1)
tele_circuit = QuantumCircuit(qr, crz, crx)
tele_circuit.draw('mpl')
```

```python
'''
Here, Alice has q1 and Bob has q2 in entabgled state (entabgled by Telemon)
Alice wishes to send the qubit q0 to Bob using Teleportation.
'''
bell_pair(tele_circuit, qr[1], qr[2])
tele_circuit.draw('mpl')
```

```python
# Add a barrier between steps
tele_circuit.barrier()
tele_circuit.draw('mpl')
```

```python
tele_circuit.cx(qr[0], qr[1])
tele_circuit.h(qr[0])
tele_circuit.draw('mpl')
```

```python
tele_circuit.barrier()
tele_circuit.draw('mpl')
```

```python
def measure_alice_qubits(circuit, qr1, qr2):
    circuit.measure(qr1, 0)
    circuit.measure(qr2, 1)
```

```python
measure_alice_qubits(tele_circuit, qr[0], qr[1])
tele_circuit.barrier()
tele_circuit.draw('mpl')
```

```python
def bob_gates(circuit, qr, crz, crx):
    circuit.x(qr).c_if(crx, 1)
    circuit.z(qr).c_if(crz, 1)
```

```python
bob_gates(tele_circuit, qr[2], crz, crx)
tele_circuit.draw('mpl')
```
