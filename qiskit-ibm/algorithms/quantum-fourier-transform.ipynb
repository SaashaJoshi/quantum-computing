```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from math import pi
```

```python
def qft(n):
    circuit = QuantumCircuit(n)
    
    for qbit in range(n):
        circuit.h(qbit)
        for next_qbit in range(qbit + 1, n):
            circuit.cu1(pi / (2**(next_qbit - qbit)), next_qbit, qbit)
        
    return circuit
```

```python
circuit = qft(5)
circuit.draw('mpl')
```

```python
def qft_dagger(n):
    circuit = QuantumCircuit(n)

    for qbit in range(n//2):
        circuit.swap(qbit, n-qbit-1)
        
#     circuit.barrier()
    
    for qbit in range(n):
        for next_qbit in range(qbit):
            circuit.cu1(-pi/(2**(qbit - next_qbit)), next_qbit, qbit)
        circuit.h(qbit)
        
    return circuit
```

```python
circuit = qft_dagger(5)
circuit.draw('mpl')
```
