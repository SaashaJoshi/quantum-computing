```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere
from qiskit.quantum_info import Statevector
```

```python
sv = Statevector.from_label('00')

# Bell State |Î¨+>
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
# qc.draw('mpl')
new_sv = sv.evolve(qc)
print(new_sv)
plot_state_qsphere(new_sv.data)
```

```python
sv = Statevector.from_label('00')

# Bell State |Î¨->
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.z(0)
# qc.draw('mpl')
new_sv = sv.evolve(qc)
print(new_sv)
plot_state_qsphere(new_sv.data)
```
