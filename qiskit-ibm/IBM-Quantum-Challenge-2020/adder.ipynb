```python
from qiskit import *
from qiskit.visualization import *
```

```python
qc = QuantumCircuit(11, 11)

# qc.reset(range(4))
qc.h(range(8))

qc.cx(0, 8)
qc.ccx(1, 8, 9)
qc.cx(1, 8)
qc.ccx(2, 8, 9)
qc.cx(2, 8)
qc.mcx([3, 8, 9], 10)
qc.ccx(3, 8, 9)
qc.cx(3, 8)
qc.mcx([4, 8, 9], 10)
qc.ccx(4, 8, 9)
qc.cx(4, 8)
qc.mcx([5, 8, 9], 10)
qc.ccx(5, 8, 9)
qc.cx(5, 8)
qc.mcx([6, 8, 9], 10)
qc.ccx(6, 8, 9)
qc.cx(6, 8)
qc.mcx([7, 8, 9], 10)
qc.ccx(7, 8, 9)
qc.cx(7, 8)    # 8 is S.
qc.barrier()

qc.barrier()

# control_qubits = [qubits for qubits in range(8)]
# qc.mcx(control_qubits, 11)    # If number is 8
qc.barrier()

qc.measure([8, 9, 10, 7, 6, 5, 4, 3, 2, 1, 0], range(11))
# qc.barrier()

qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
results = execute(qc, backend, shots = 8192).result()
counts = results.get_counts()
print(counts)
plot_histogram(counts)
```
