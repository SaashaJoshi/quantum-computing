```python
from qiskit import *
```

```python
qc = QuantumCircuit(4, 2)

qc.reset(range(4))
qc.x(0)
qc.x(2)

# Sum
qc.cx(0, 3)
qc.cx(1, 3)
qc.cx(2, 3)
qc.barrier()

# Carry
qc.cx(0, 1)
qc.cx(0, 2)
qc.ccx(1, 2, 0)
qc.barrier()
qc.measure([3, 0], [0, 1])

# qc.barrier()

qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
execution = execute(experiments = qc, backend = backend, shots = 1024)    
results = execution.result()
counts = results.get_counts()
print(counts)
```
